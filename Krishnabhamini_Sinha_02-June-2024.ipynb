## 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

```python
df = pd.read_excel('Zingaro_Company.xlsx')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.distplot(df['Unpolished'], kde=True)
plt.show()
```

$H_0$ : $\mu$ = 150

$H_a$ : $\mu$ < 150

$\alpha$ = 0.05

```python
print('The sample size for this problem is:', len(df['Unpolished']))
print('The mean hardness index for unpolished stones is:', df['Unpolished'].mean())
```

```python
from scipy import stats
from scipy.stats import ttest_1samp, ttest_ind
```

```python
t_statistic, p_value = ttest_1samp(df['Unpolished'], 150, alternative='less')
print('One sample t-test p-value:', p_value)
```

p_value < $\alpha$

We reject $H_0$. 


## 3.2 Is the mean hardness of the polished and unpolished stones the same?

$H_0$ : $\mu1$ = $\mu2$

$H_a$: $\mu1$ $\neq$ $\mu2$

$\alpha$ = 0.05

We will conduct a Two Sample t-test as we are comparing two indpendent samples. 

```python
t_statistic, p_value = ttest_ind(df['Unpolished'], df['Treated and Polished'])
print('t statistic', t_statistic)
print('p-value', p_value)
```

p-value < $\alpha$

We reject $H_0$

Hypotheses for Alloy 2:

## Checking Assumptions

**Shapiro-Wilk's test: **

```python
w, p_value = stats.shapiro(alloy1['Dentist'])
print('The p-value is:', p_value)
```

p-value < $\alpha$

```python
w, p_value = stats.shapiro(alloy2['Dentist'])
print('The p-value is:', p_value)
```

So, we reject $H_0$ i.e., distribution is not normal for Alloy2

** Levene's Test**

```python
from scipy.stats import levene

w, p_value = levene(data['Response'] [data['Dentist']==1],
                    data['Response'] [data['Dentist']==2],
                    data['Response'] [data['Dentist']==3],
                    data['Response'] [data['Dentist']==4],
                    data['Response'] [data['Dentist']==5])

print('p-value is:', p_value)
```

p-value < $\alpha$

So, reject the null hypothesis. Variance of at least one sample is different.
